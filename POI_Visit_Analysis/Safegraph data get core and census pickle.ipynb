```python
data_feb_cbgs = data_feb_harris_geo[data_feb_harris_geo.visitor_work_cbgs.notnull()]
data_feb_cbgs = data_feb_cbgs[data_feb_cbgs.visitor_home_cbgs.notnull()]
data_feb_cbgs = data_feb_cbgs[['safegraph_place_id', 'visitor_work_cbgs', 'visitor_home_cbgs', 'latitude', 'longitude']]
```

```python
data_feb_cbgs
```

```python
filter1 = [len(x)>2 for x in data_feb_cbgs['visitor_work_cbgs']]

data_feb_cbgs['filter1'] = filter1

data_feb_cbgs = data_feb_cbgs[data_feb_cbgs.filter1 ==True]
```

```python
filter2 = [len(x)>2 for x in data_feb_cbgs['visitor_home_cbgs']]

data_feb_cbgs['filter2'] = filter2

data_feb_cbgs = data_feb_cbgs[data_feb_cbgs.filter2 ==True]
```

```python
def Load_json(arg1):
    return json.loads(arg1)
```

```python
data_feb_cbgs['visitor_home_cbgs'] = data_feb_cbgs['visitor_home_cbgs'].apply(Load_json)
data_feb_cbgs['visitor_work_cbgs'] = data_feb_cbgs['visitor_work_cbgs'].apply(Load_json)
```

```python
data_feb_cbgs = data_feb_cbgs.drop(['filter1','filter2'], axis=1)
```

```python
def Map_cbg_network(row):
    return_list = []
    for (key,value) in row['visitor_work_cbgs'].items():
        return_list.append([row['safegraph_place_id'], row['longitude'], row['latitude'], key, value])
    for (key,value) in row['visitor_home_cbgs'].items():
        return_list.append([row['safegraph_place_id'], row['longitude'], row['latitude'], key, value])
    return return_list
```

```python
movement = data_feb_cbgs.apply(Map_cbg_network, axis=1)
```

```python
movement_dataframe = movement.to_frame()
```

```python
movement_dataframe.columns = ['structure']
```

```python
data_filter = [x[0] for x in movement_dataframe['structure']]
```

```python
movement_dataframe = pd.DataFrame(data_filter, columns=["safegraph_place_id'", "longitude", "latitude", "cbg", "visits"])
```

```python
movement_dataframe['cbg'] = movement_dataframe['cbg'].apply(int)
```
